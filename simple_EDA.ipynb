#### Statistical Summary
Most variables (e.g., LB, AC, FM, UC, ASTV, MSTV, etc.) are continuous and show different ranges of fetal heart metrics.

NSP is categorical with three classes:

1 = Normal -> majority class.

2 = Suspect -> smaller share.

3 = Pathologic -> least frequent.

Tendency was cleaned and now only has values -1, 0, 1 with mode = 0 (Stable) dominating.

There is more data variation in DS and DP.

#### Visualization Insights

Histograms -> Most numerical variables are skewed (not perfectly normal). Some like LB (baseline fetal heart rate) show clustering.

Boxplots -> Several variables (e.g., UC, FM, ASTV) have outliers, which may represent abnormal fetal conditions.

NSP Distribution -> Imbalanced target: most fetuses are Normal, fewer Suspect, and very few Pathologic.

Tendency Distribution -> Majority of cases are Stable (0).

Correlation Heatmap ->

Strong correlations exist between variability measures (e.g., ASTV, MSTV, ALTV, MLTV).

Baseline FHR (LB) is weakly correlated with most other features.

#### Patterns & Insights

Normal vs. Suspect/Pathologic:
Suspect and Pathologic cases show higher abnormal variability (ASTV, ALTV) compared to Normal.

Outliers in decelerations (DL, DS, DP) often align with Pathologic outcomes.

Tendency being mostly stable suggests only a minority of cases show trends (increasing/decreasing).
